```python
correlation = dados.corr()
```

```python
#valorSolicitado
#valorAprovado

important_features = correlation.loc[(correlation.index == "valorSolicitado") | (correlation.index == "valorAprovado")]
others = correlation.loc[(correlation.index != "valorSolicitado") | (correlation.index != "valorAprovado")]

correlation_ordened = pd.concat([others,important_features])
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
fig, ax = plt.subplots(figsize=(10,10))


plot = sn.heatmap(correlation_ordened, annot = True, fmt=".1f", linewidths=2,annot_kws={'size': 8},ax=ax)
plot
```
